```python
# Compute Key Performance Indicators (KPIs)

kpis = {}

# 1. Win Rate (Home & Away Teams)
total_matches = len(df)
home_wins = len(df[df['FTR'] == 'H'])
away_wins = len(df[df['FTR'] == 'A'])
draws = len(df[df['FTR'] == 'D'])

kpis["Home Win Rate"] = round((home_wins / total_matches) * 100, 2)
kpis["Away Win Rate"] = round((away_wins / total_matches) * 100, 2)
kpis["Draw Rate"] = round((draws / total_matches) * 100, 2)

# 2. Average Goals per Match
kpis["Avg Goals per Match"] = round(np.mean(df['FTHG'] + df['FTAG']), 2)

# 3. Shots-to-Goals Conversion Rate (Home & Away)
home_conversion_rate = np.mean(df['FTHG'] / (df['HS'] + 1))  # Adding 1 to avoid division by zero
away_conversion_rate = np.mean(df['FTAG'] / (df['AS'] + 1))

kpis["Home Shots-to-Goal Conversion Rate (%)"] = round(home_conversion_rate * 100, 2)
kpis["Away Shots-to-Goal Conversion Rate (%)"] = round(away_conversion_rate * 100, 2)

# 4. Clean Sheet Percentage (Games where no goals were conceded)
home_clean_sheets = len(df[df['FTAG'] == 0])
away_clean_sheets = len(df[df['FTHG'] == 0])

kpis["Home Clean Sheet Percentage"] = round((home_clean_sheets / total_matches) * 100, 2)
kpis["Away Clean Sheet Percentage"] = round((away_clean_sheets / total_matches) * 100, 2)

# 5. Betting Accuracy (How often odds predict correct outcomes)
df['Predicted Result'] = df[['AvgH', 'AvgD', 'AvgA']].idxmin(axis=1).map({'AvgH': 'H', 'AvgD': 'D', 'AvgA': 'A'})
correct_predictions = np.sum(df['Predicted Result'] == df['FTR'])
kpis["Betting Odds Prediction Accuracy (%)"] = round((correct_predictions / total_matches) * 100, 2)

# Display the computed KPIs
kpis
```

```python
# Import necessary libraries for logistic regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Encode target variable (FTR: H -> 0, D -> 1, A -> 2)
label_encoder = LabelEncoder()
df['FTR_Encoded'] = label_encoder.fit_transform(df['FTR'])

# Select features for prediction
features = ['FTHG', 'FTAG', 'HTHG', 'HTAG', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR']
X = df[features]
y = df['FTR_Encoded']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# Train Logistic Regression Model
log_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=500)
log_reg.fit(X_train, y_train)

# Make predictions
y_pred = log_reg.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)

# Display accuracy and classification report
accuracy, conf_matrix, class_report
```

```python

# Select a sample match (taking the first row of the test set)
sample_match = X_test[0].reshape(1, -1)  # Reshape to maintain array shape for prediction

# Predict the outcome for the sample match
predicted_class = log_reg.predict(sample_match)[0]

# Convert prediction back to original label (H, D, A)
predicted_outcome = label_encoder.inverse_transform([predicted_class])[0]

print(f"Predicted Match Outcome: {predicted_outcome}")




```
